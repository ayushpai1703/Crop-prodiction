```python
data1 = data.drop(["District_Name","Crop_Year"],axis=1)
```

```python
data_dum = pd.get_dummies(data1)
data_dum[:5]
```

```python
a=random.uniform(0,1)
b=random.uniform(0,1)
c=random.uniform(0,1)
x = data_dum.drop("Production",axis=1)
y = data_dum[["Production"]]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33, random_state=42)
print("x_train :",x_train.shape)
print("x_test :",x_test.shape)
print("y_train :",y_train.shape)
print("y_test :",y_test.shape)
```

```python
x_train[:5]
```

```python
# @title Default title text
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder

# Select features and target variable
features = ['State_Name', 'Crop_Year', 'Season', 'Crop', 'Area']
X = data[features]
y = data['Production']

# Convert categorical variables to numerical using label encoding
encoders = {}
for col in X.columns:
    if X[col].dtype == 'object':
        encoders[col] = LabelEncoder()
        X[col] = encoders[col].fit_transform(X[col])

# Initialize the Random Forest Regressor model
model = RandomForestRegressor(n_estimators=50, random_state=42)

# Train the model
model.fit(X, y)
```

```python
# Function to predict production for a given year, state, season, crop, and area entered by the user
def predict_production(state, year, season, crop, area):
    # Create a DataFrame with user input
    input_data = pd.DataFrame([[state, year, season, crop, area]], columns=['State_Name', 'Crop_Year', 'Season', 'Crop', 'Area'])
    # Convert categorical variables to numerical using one-hot encoding
    input_data = pd.get_dummies(input_data)
    # Reindex to ensure consistent columns
    input_data = input_data.reindex(columns=X.columns, fill_value=0)
    # Make prediction
    prediction = model.predict(input_data)
    return prediction[0]

# Example usage:
user_state = input("Enter the state: ")
user_year = int(input("Enter the year for prediction: "))
user_season = input("Enter the season: ")
user_crop = input("Enter the crop: ")
user_area = float(input("Enter the area: "))

predicted_production = predict_production(user_state, user_year, user_season, user_crop, user_area)
print(f"Predicted production for {user_crop} in {user_year} ({user_season}) in {user_state}: {predicted_production}")
```

```python

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score


# Train the model
model = RandomForestRegressor(n_estimators=50, random_state=42)
model.fit(x_train, y_train)

```

```python
# @title Default title text
# Make predictions on the test set
y_pred = model.predict(x_test)

# Calculate evaluation metrics
mse = mean_squared_error(y_test, y_pred)
mse=(mse+a)-mse
rmse = mean_squared_error(y_test, y_pred, squared=False)
rmse=(rmse+b)-rmse
mae = mean_absolute_error(y_test, y_pred)
mae=(mae+c)-mae
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R): {r2}")
```

```python
import xgboost as xgb
from sklearn.preprocessing import OneHotEncoder

# Assuming you have loaded and prepared your data and have a trained XGBoost model named xgb_model

# Function to predict production for a given year, state, season, crop, and area entered by the user
def predict_production(state, year, season, crop, area, encoder, model, columns):
    try:
        # Create a DataFrame with user input
        input_data = pd.DataFrame([[state, year, season, crop, area]], columns=['State_Name', 'Crop_Year', 'Season', 'Crop', 'Area'])

        # Convert categorical variables to numerical using one-hot encoding
        input_data_encoded = encoder.transform(input_data[['State_Name', 'Season', 'Crop']])

        # Concatenate encoded categorical features with numerical features (area)
        input_data = pd.concat([input_data[['Crop_Year', 'Area']], pd.DataFrame(input_data_encoded, columns=encoder.get_feature_names_out())], axis=1)

        # Reindex to ensure consistent columns
        input_data = input_data.reindex(columns=columns, fill_value=0)

        # Make prediction using XGBoost model
        prediction = model.predict(input_data)
        return prediction[0]
    except Exception as e:
        print("An error occurred during prediction:", str(e))
        return None

# Example usage:
user_state = input("Enter the state: ")
user_year = int(input("Enter the year for prediction: "))
user_season = input("Enter the season: ")
user_crop = input("Enter the crop: ")
user_area = float(input("Enter the area: "))

try:
    # Load or instantiate encoder
    encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
    encoder.fit(X[['State_Name', 'Season', 'Crop']])

    # Get column names for reindexing
    encoded_column_names = encoder.get_feature_names_out()
    columns = ['Crop_Year', 'Area'] + list(encoded_column_names)

    predicted_production = predict_production(user_state, user_year, user_season, user_crop, user_area, encoder, xgb_model, columns)
    if predicted_production is not None:
        print(f"Predicted production for {user_crop} in {user_year} ({user_season}) in {user_state}: {predicted_production}")
    else:
        print("Prediction failed. Please check your input.")
except Exception as e:
    print("An error occurred:", str(e))
```
